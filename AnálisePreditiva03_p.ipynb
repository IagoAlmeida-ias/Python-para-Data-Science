```python
# Pacotes de manipulação de Dados

import numpy as np
import pandas as pd

# Pacotes graficos

import matplotlib.pyplot as plt
import seaborn as sns

# Pacotes de modelagem

import statsmodels.api as sm
```

```python
# Importação da base de dados

df = pd.read_csv('base_funcionarios_v3.csv')
```

```python
colunas = ['id', 'Salario', 'Anos_Educ_Superior', 'Tempo_Empresa',
       'Tempo_de_Carreira']
df.columns = colunas
```

```python
df = df.set_index('id')
```

# Complementando - Análise Exploratória dos Dados

```python
# Gráfico de Dispersão por Superior

sns.scatterplot(x=df['Anos_Educ_Superior'],
                y=df['Salario']);
```

```python
# Correlação linear Persen

df.corr()
```

```python
# Grafico HeatMap

sns.heatmap(df.corr(), cmap='RdBu_r', annot=True, vmin=-1, vmax=1);
```

```python
# Variável resposta

y= df['Salario']

# Variável explicativa

df['intercepto'] = 1
x = df[ ['intercepto', 'Anos_Educ_Superior', 'Tempo_Empresa', 'Tempo_de_Carreira'] ]
```

```python
# Ajuste do Modelo

modelo = sm.OLS(y,x)
res = modelo.fit()
```

```python
# Resultado do ajuste do modelo

print(res.summary())
```
