```python
colunas = ['id', 'produto', 'categoria', 'preco', 'quantidade', 'data_venda',
       'vendedor']

df.columns = colunas
```

```python
df['preco'] = df['preco'].str.replace("'", "")
```

```python
df['preco'] = df['preco'].astype({'preco':'float64'})
```

```python
df['data_venda'] = pd.to_datetime(df['data_venda'], format='%Y-%m-%d')
```

```python
df['produto'].fillna('Produto Não Indentificado', inplace=True)
df['categoria'].fillna('Categoria Não Indentificada', inplace=True)
df['vendedor'].fillna('Vendedor Não Indentificado', inplace=True)

df['data_venda'].ffill(inplace=True)
```

```python
df['preco_cus'] = df['preco'] - (df['preco'] * 0.70 )
df['valor_final'] = df['preco'] * df['quantidade']
df['margem_lucro'] = df['valor_final'] - (df['quantidade'] * df['preco_cus'])
df['perc_margem_lucro'] = df['margem_lucro'] / df['preco']
```

### Análise de Rentabilidade por Produto

```python
df.groupby('produto').agg(
    {
       'margem_lucro':'sum',
       'perc_margem_lucro':'mean'
    }
).sort_values(by='margem_lucro', ascending=False).round(2)
```

### Análise de Rentabilidade por Vendedor

```python
df.groupby('vendedor').agg(
    {
        'margem_lucro': 'sum',
        'perc_margem_lucro': 'mean'
    }
).sort_values(by='margem_lucro', ascending=False).round(2)
```

# Rentabilidade por Categoria

```python
df.groupby('categoria').agg(
    {
        'margem_lucro': 'sum',
        'perc_margem_lucro': 'mean'
    }
).sort_values(by='margem_lucro', ascending=False).round(2)
```
