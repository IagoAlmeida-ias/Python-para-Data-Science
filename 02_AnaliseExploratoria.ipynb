# Análise Exploratória
Vamos aprender a como utilizar as principais funcianalidades do Pandas
para fazer uma análise explorativa inicial dos dados.

Objetivos:
- Enternder melhor os dados
- Identificar problemas de dados
- Verificar se carregou os dados corretamente
- Obter as primeiras estatíticas dos dados

```python
# Importando a biblioteca do pandas
import pandas as pd
```

```python
# Carregando Arquivo .CSV
df = pd.read_csv('base_vendas.csv')
```

```python
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
df.shape
```

```python
# Primeiras Estatíticas da Base de Dados
# Valores Numéricos

df.describe().round(1)
```

```python
# Primeiras Estatíticas da Base de Dados
# Valores Texto

df.describe(include=object).round(1)
```

```python
# Indetificando Valores Faltantes ou Valores Nulos
# isnull()
# isna()

df.isnull()
```

```python
df.isnull().sum()
```

```python
# Verificando Valores Duplicados

print(df.duplicated().sum())
```

```python
# Verificando Valores Duplicados em uma coluna Específica

print(df['Nome Funcionario'].duplicated().sum())
```
