# Filtrando Dados do DataFreme

```python
# Teste Lógicos
# Exemplo com uma condição

filtro_vf = df['valor_final'] < 1000
df[filtro_vf]
```

```python
# Teste Lógicos
# Exemplo com múltiplas condições

filtro_vf = df['valor_final'] < 1000
filtro_lj = df['nome_loja'] == 'Filial SP'
df[filtro_vf & filtro_lj]
```

#  Filtrando registro do DataFreme com o metodo query

```python
# Exemplo com uma condição

df.query('valor_final < 1000')
```

```python
# Exemplo com múltiplas condições

df.query("valor_final < 1000 and nome_loja == 'Filial SP'")
```

```python
# Procura uma Loja

df.query("nome_loja == 'Loja Matriz'")
```

```python
# Procurar dois fatores

df_fil =  df.query("nome_loja == 'Loja Matriz' or descricao_produto == 'Fritadeira elétrica'")
```

```python
df_fil[colunas].sum()
```

```python
# Utilizando o index para filtrar o DataFreme

filtro_ano = df.index.year == 2020
filtro_mes = df.index.month == 2
filtro_fum = df['nome_funcionario'] == 'Jorge Mateus'

df[filtro_ano & filtro_mes & filtro_fum]
```

# Agrupando dados com o metodo groupby()

```python
# Exemplo 01 : Agrupamento simples com uma Coluna

df.groupby('nome_funcionario')['valor_final'].sum()
```

```python
# Exemplo 02 : Agrupamento com mais de uma Coluna

df.groupby(['nome_funcionario', 'descricao_produto'])[ ['valor_final', 'preco_custo', 'quantidade'] ].sum()
```
