```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('base_bom_pagador.csv')
```

```python
colunas = ['bom_pagador', 'renda_mensal', 'idade', 'estado_civil',
       'possui_cartao_credito', 'escolaridade', 'tempo_emprego', 'qtd_filhos',
       'score_credito', 'historico_inadimplencia', 'utiliza_limite_cartao',
       'valor_emprestimos', 'tipo_residencia', 'trabalha_formalmente']

df.columns = colunas
```

```python
df.duplicated().sum()
```

```python
df.isnull().sum()
```

```python
df.describe().round()
```

```python
df.query("bom_pagador == 'Não'")
```

```python
df.query("score_credito > 600 and bom_pagador == 'Não'")
```

```python
df.query("score_credito < 400 and bom_pagador == 'Sim'")
```

```python
df.query("idade < 28 and valor_emprestimos > 20000")
```

```python
df.query("idade >= 18 and idade < 25 and tempo_emprego > 10")
```

```python
df_etc = df.groupby('estado_civil')['valor_emprestimos'].sum().sort_values(ascending=False)
```

```python
df_etc.plot(kind='bar',figsize=(10,5),y='total_vendas');
```

```python
df_e = df.groupby('escolaridade')['valor_emprestimos'].sum().sort_values(ascending=True)
```

```python
df_e.plot(kind='barh',figsize=(10,5),y='total_vendas');
```

```python
df_i = df.groupby('qtd_filhos')['valor_emprestimos'].sum()
```

```python
df_i.plot(kind='line', figsize=(13,5),linestyle='--', color='red', marker='o');
```

```python
df_ren = df.groupby('tipo_residencia')['renda_mensal'].sum().sort_values(ascending=False)
```

```python
df_ren.plot(kind='bar', figsize=(13,5), color='blue');
```
