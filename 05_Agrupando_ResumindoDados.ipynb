```python
# Criando margem de lucro

df['margem_lucro'] = df.valor_final - (df.preco_custo * df.quantidade)

# Calculo % Margem de lucro
df['perc_margem_lucro'] = df.margem_lucro / df.valor_final

df.head()
```

# Análise Rentabilidade
-- Categoria_produto ||| ML (soma)

```python
# Calcular a Marm_lucro e a perc_margem_lucro Média
#    para cada categoria de produto, ordenando-as para indentificar
#    as mais e menos lucrativas
```

```python


df.groupby('categoria')['margem_lucro'].sum().sort_values(ascending=False)
```

```python
# prof

df.groupby('categoria')['margem_lucro'].sum().round(2)
```

```python
print('### Análise de Rentabilidade por Categoria de Produto ###\n')

df.groupby('categoria').agg(
    {
        'margem_lucro':'sum',
        'perc_margem_lucro':'mean'
    }
).sort_values(by='margem_lucro', ascending=False)
```

```python
print('### Análise de Rentabilidade por Categoria de Produto ###\n')

df_renta = df.groupby('categoria').agg(
    {
        'margem_lucro':['sum', 'min', 'mean', 'max'],
        'perc_margem_lucro':'mean'
    }
)
```

```python
df_renta['perc_margem_lucro'] * 100
```

### Análise de Rentabilidade por loja

```python
# Qual loja gera mais lucro e qual tem a melhor margem percentual média.

df.groupby('nome_loja').agg(
    {
        'margem_lucro':'sum',
        'perc_margem_lucro': 'mean'
    }
).sort_values(by='margem_lucro', ascending=False)
```

### Análise de Rentabilidade por Produto

```python
# Encontrar os produtos que mais contribuem individualmente para o lucro,
#   tanto em valor absoluto quanto em percentual

df.groupby('descricao_produto').agg(
    {
        'margem_lucro':'sum',
        'perc_margem_lucro': 'mean'
    }
).sort_values(by='margem_lucro', ascending=False)
```

### Análise de Rentabilidade por Vendedor

```python
# Agrupar as vendas por nome_funcionaria para identificar quem esta contribuindo mais
#   para o lucro da empresa

df.groupby('nome_funcionario').agg(
    {
        'margem_lucro':'sum',
        'perc_margem_lucro': 'mean'
    }
).sort_values(by='margem_lucro', ascending=False)
```
