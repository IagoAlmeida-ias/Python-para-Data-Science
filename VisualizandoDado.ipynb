```python
# Pacotes de Manipulacão de Dados

import pandas as pd
import numpy as np

# Pacote de visualização de Dados

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Carregando Dataset

df = pd.read_csv('base_dados_vendas_tratada.csv', parse_dates=['dt_venda', 'dt_entrega'], index_col='dt_venda')
df.head()
```

```python
# Gráfico de Linhas
df['ano_mes'] = df.index.to_period('M')
```

```python
df
```

```python
df_vendas = df.groupby('ano_mes')['valor_final'].sum()
```

```python
#df_vendas.plot(kind='line') linha
#df_vendas.plot(kind='bar') barra

# plt.figsize()
df_vendas.plot(kind='line', figsize=(13,5),linestyle='--', color='red', marker='o')
plt.title('Vendas Mensaís', fontsize='18')
plt.xlabel('Ano/Mes', fontsize='14')
plt.ylabel('Faturamento', fontsize='14');
```

```python
df_pro = df.groupby('descricao_produto').agg(
    total_vendas=('valor_final', 'sum'),
    total_quantidade=('quantidade', 'sum')
).sort_values(by='total_vendas', ascending=False)
```

```python
df_pro.head(10).plot(kind='bar',figsize=(10,5),y='total_vendas')
plt.title('Produtos mais Vendidos', fontsize='18')
plt.xlabel('Produtos', fontsize='14')
plt.ylabel('Quantidade', fontsize='14');
```

```python
df_pro.sort_values(by='total_vendas', ascending=True, inplace=True)
df_pro.head(10).plot(kind='barh',figsize=(10,5),y='total_vendas');  # Outra forma de grafico com barra
```

```python
df_catg = df.groupby('categoria')['valor_final'].sum()
```

```python
df_catg.plot.pie()
```

```python
plt.figure(figsize=(8,8))
df_catg.plot.pie(autopct='%1.1f%%', cmap='tab20', legend=True);
```
